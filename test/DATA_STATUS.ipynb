```python
import os
import cv2
from tqdm import tqdm
from google.colab.patches import cv2_imshow
import pandas as pd

data_path = '/content/drive/MyDrive/PDC_2022/data'
data_dir_list = [folders for folders in os.listdir(data_path) if "(" not in folders]
print(data_dir_list)
print("===========================\n")
count = 0
for i in range(len(data_dir_list)): #팀원별
  data_list = os.listdir(data_path + "/" + data_dir_list[i])
  data_list = sorted(data_list)
  print(data_dir_list[i])
  for j in range(len(data_list)): #레이블별
    data_access_path_list = os.listdir(data_path + "/" + data_dir_list[i] + "/" +data_list[j])
    print(data_list[j])
    img_path_list = []
    for k in tqdm(range(80)): #파일별
      if not data_access_path_list:
        break
      if k >= len(data_access_path_list):
        count += k
        break
      else:
        img_path = data_path + "/" + data_dir_list[i] + "/" +data_list[j] + "/" + data_access_path_list[k]
        img_path_list.append(img_path)
  print("\n===========================")
print("전체 진행도 :{}/2,800, {:.2f}%".format(count,count/2800*100))
print("===========================")
```
